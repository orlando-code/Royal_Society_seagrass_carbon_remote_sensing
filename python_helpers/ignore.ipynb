```python
import pandas as pd

df = pd.read_csv(
    "/Users/rt582/Desktop/Royal_Society_seagrass_carbon_remote_sensing-1/output/pixel_grouped_51-53-56-73-77/across_seeds_summary.csv"
)

# Concatenate relevant column pairs (e.g. mean_mean_rmse and sd_mean_rmse) with formatting: #.#### (#.####)
# Assumes the structure: model, mean_mean_rmse, sd_mean_rmse, mean_mean_r2, sd_mean_r2, ... etc

# drop XGB row
df = df[df["model"] != "XGB"]
df.set_index("model", inplace=True)

# reorder df
df = df.reindex(["GPR", "GAM", "LR"])

out_df = pd.DataFrame()
# Start from column 1 (skip model), step by 2 for mean/sd pairs
for i in range(0, len(df.columns), 2):
    mean_col = df.columns[i]
    if i + 1 < len(df.columns):
        sd_col = df.columns[i + 1]
        formatted = df.apply(
            lambda row: f"{row[mean_col]:.4f} ({row[sd_col]:.4f})", axis=1
        )

        out_df[mean_col] = formatted
(out_df.T).to_csv(
    "/Users/rt582/Desktop/Royal_Society_seagrass_carbon_remote_sensing-1/output/pixel_grouped_51-53-56-73-77/summary_table.csv"
)
```
